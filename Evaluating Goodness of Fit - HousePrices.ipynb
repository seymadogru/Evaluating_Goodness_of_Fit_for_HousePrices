```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
house_prices_df = pd.read_csv("house_prices_train.csv")
house_prices_df.head()
```

```python
new_features = house_prices_df[["YearBuilt","LotArea","TotRmsAbvGrd","FullBath","MoSold","HalfBath","GarageCars","SalePrice"]]
Y = new_features['SalePrice']
X = new_features[['YearBuilt','LotArea','MoSold','TotRmsAbvGrd','FullBath','HalfBath','GarageCars']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)


X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

**Ev fiyatları modelinizi tekrar çalıştırın ve F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC kullanarak modelinizin uygunluğunu değerlendirin.**

F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC değerlerine baktığımızda F-Testi bir miktar düşük çıktı. Önceki ödevden elde ettiğim kanıya göre alakasız feature'ları çıkardığımda bu değerin yükseldiğini gördüm.

R-Kare değeri %70 in üzerine çıktığında model bizim için iyi diyebiliriz. Ancak burada %60 da kaldı. Orta derecede kaliteli bir modelimiz var diyebiliriz. Yani modelimizde bağımsız değişken bağımlı değişken hakkında %60 oranında tahminde bulunabiliyor.

Ayarlanmış R-kare değerinin R-kare ile birbirlerine çok yakın olduğunu görüyoruz. Sadece R-kare değeri 0.604 iken o 0.602 de kalmış. ayarlanmış R-kare değeri bizi korele değer varsa cezalandırır ve yavaş yükselir. Buradan da veri setimizde ufak bir korelasyon vardır diyebiliriz. Ancak bu değer çok da yüksek değildir çünkü  R-kare değerine hayli yakın çıkmış.

AIC ve BIC değerleri ne kadar düşük olursa o kadar iyidir bizim için. Muhtemelen buradaki p value'ları 0.05 den büyük olan değişkenleri veri setinden çıkarırsak bu değerler daha da çok düşecektir.

**Modelinizin tatmin edici olduğunu düşünüyor musunuz? Açıklayın**

Modelin tatmin edici olduğunu düşünmüyorum. 80 civarı feature varken, mutlaka aralarında hedef değişkenimizi daha iyi tahmin etmemize yarayacak bir değişken bulabiliriz. Bu R-kare değeri daha da yükselebilir. Yüzde 70in üzerinde çıkmaya başladığında bizim için iyidir diyebiliriz.

**Modelinizin uygunluğunu iyileştirmek için, bazı değişkenleri ekleyerek veya kaldırarak farklı model özelliklerini deneyin.**

OverallQual: Rates the overall material and finish of the house (Evin genel malzemesini ve bitişini değerlendirmesi)

OverallCond: Rates the overall condition of the house (Evin genel durumunun değerlendirilmesi)

1stFlrSF: First Floor square feet  (1. katın fit karesi)
 
2ndFlrSF: Second floor square feet (2. katın fit karesi)

GarageArea: Size of garage in square feet (Garajın fit karesi)


```python
new_features = house_prices_df[['YearBuilt','LotArea','OverallQual','TotRmsAbvGrd','FullBath','OverallCond','GarageCars','1stFlrSF','2ndFlrSF','GarageArea',"SalePrice"]]
Y = new_features['SalePrice']
X = new_features[['YearBuilt','LotArea','OverallQual','TotRmsAbvGrd','FullBath','OverallCond','GarageCars','1stFlrSF','2ndFlrSF','GarageArea']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)


X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```
